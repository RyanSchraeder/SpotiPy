```python
import graphviz
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd 
import seaborn as sns

## ML Dependencies. Using Decision Tree, Random Forest 

from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, export_graphviz
```

```python
data = pd.read_csv("./Resources/data.csv")
data.head(10)
```

```python
## Identify metric groups for sole dataframe
# valence, valence, Danceability, Instrumentalness

data_df = data[['valence', 
                'energy',
                'danceability', 
                'instrumentalness',
                'target']]

valence = pd.DataFrame(data_df[['valence','target']])

positive_valence = valence [valence['target'] == 1]['valence']

negative_valence = valence[valence['target'] == 0]['valence']


## HISTOGRAM DISPLAY
# Label axes

plt.figure(figsize=(14, 8))
sns.set()

## RESET SNS TO STYLE GRAPH
# sns.set_style('white')
# colors = [, 10]
# palette = sns.diverging_palette(240, 10, n=9)
# sns.set_palette(palette)

positive_valence.hist(alpha=.5, bins=40, label='positive_valence')
negative_valence.hist(alpha=.5, bins=40, label='negative_valence')

plt.title('Relationship of Valence & Preference')
plt.legend()
```

```python
## ENERGY 

energy = pd.DataFrame(data_df[['energy','target']])

positive_energy = energy [energy['target'] == 1]['energy']

negative_energy = energy[energy['target'] == 0]['energy']

## HISTOGRAM DISPLAY
# Label axes

plt.figure(figsize=(14, 8))
sns.set()

## RESET SNS TO STYLE GRAPH
# sns.set_style('white')
# colors = [, 10]
# palette = sns.diverging_palette(240, 10, n=9)
# sns.set_palette(palette)

positive_energy.hist(alpha=.5, bins=40, label='positive_energy')
negative_energy.hist(alpha=.5, bins=40, label='negative_energy')

plt.title('Relationship of energy & Preference')
plt.legend()
```

```python
## DANCEABILITY 

danceability = pd.DataFrame(data_df[['danceability','target']])

positive_danceability = danceability [danceability['target'] == 1]['danceability']

negative_danceability = danceability[danceability['target'] == 0]['danceability']

## HISTOGRAM DISPLAY
# Label axes

plt.figure(figsize=(14, 8))
sns.set()

## RESET SNS TO STYLE GRAPH
# sns.set_style('white')
# colors = [, 10]
# palette = sns.diverging_palette(240, 10, n=9)
# sns.set_palette(palette)

positive_danceability.hist(alpha=.5, bins=40, label='positive_danceability')
negative_danceability.hist(alpha=.5, bins=40, label='negative_danceability')

plt.title('Relationship of danceability & Preference')
plt.legend()
```
