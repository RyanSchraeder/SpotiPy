```python
data = pd.read_csv("./Resources/data.csv").drop(columns='Unnamed: 0')
data.dropna(how='any')
data.head(10)
```

## Data Understanding 
> The weights of each attribute are based upon the presence of that attribute in a song, as well as its relationship to a positive or negative reaction of the user. 

- The user will simply provide a sentiment as a reaction to a song which will be graded on a boolean scale, where "0" is a negative reaction to the song and "1" is a positive reaction to the song in the 'Target' column.

```python
## Identify metric groups for one dataframe
# Valence, Energy, Danceability, Instrumentalness

data_df = data[['valence', 
                'energy',
                'danceability', 
                'instrumentalness',
                'target']]
```

## Valence Relationship

```python

valence = pd.DataFrame(data_df[['valence','target']])

positive_valence = valence [valence['target'] == 1]['valence']

negative_valence = valence[valence['target'] == 0]['valence']


## HISTOGRAM DISPLAY
# Label axes

plt.figure(figsize=(14, 8))
sns.set()

## RESET SNS TO STYLE GRAPH
sns.set_style('white')
# Diverging Palette if range allows
# palette = sns.diverging_palette(240, 5, n=50)
# sns.set_palette(palette)

positive_valence.hist(alpha=.5, bins=40, label='Positive Sentiment Provided')
negative_valence.hist(alpha=.5, bins=40, label='Negative Sentiment Provided')

plt.title('Relationship of Valence & Preference')
plt.ylabel('Song Count Rated')
plt.xlabel('Valence Percentage (Song Content)')
plt.legend()
```

```python
## ENERGY 

energy = pd.DataFrame(data_df[['energy','target']])

positive_energy = energy [energy['target'] == 1]['energy']

negative_energy = energy[energy['target'] == 0]['energy']

## HISTOGRAM DISPLAY
# Label axes

plt.figure(figsize=(14, 8))
sns.set()

## RESET SNS TO STYLE GRAPH
sns.set_style('white')
# Diverging Palette if range allows
# palette = sns.diverging_palette(240, 5, n=50)
# sns.set_palette(palette)

positive_energy.hist(alpha=.5, bins=40, label='Positive Sentiment Provided')
negative_energy.hist(alpha=.5, bins=40, label='Positive Sentiment Provided')

plt.title('Relationship of Energy & Preference')
plt.xlabel("Energy Percentage (Song Content)")
plt.ylabel("Song Count Rated")
plt.legend()
```
